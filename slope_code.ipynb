```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.loadtxt('ball_med_reversed_2.csv', skiprows=15, delimiter=',')
```

```python
t = data[:, 0]
V = data[:, 1]

plt.plot(t, V, label='Voltage vs Time')
plt.xlim(0, 1)
plt.ylim(0.5, 3)
plt.xlabel('Time (s)')
plt.ylabel('Voltage (V)')
plt.title('Voltage vs Time Plot')
plt.grid()
plt.legend()
```
