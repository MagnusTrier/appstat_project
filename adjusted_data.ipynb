```python
#weighted measurements form floor

floor_bottom_l = np.array([data[0]*0.001 for data in floor_to_bottom])
floor_bottom_err = np.array([data[1]*0.001 for data in floor_to_bottom])
floor_top_l = np.array([data[0]*0.001 for data in floor_to_top])
floor_top_err = np.array([data[1]*0.001 for data in floor_to_top])

floor_bottom_mean_weighted = np.average(floor_bottom_l, weights = 1.0/floor_bottom_err**2)
error_floor_bottom_weighted = np.sqrt(1.0/np.sum(1.0/np.square(floor_bottom_err)))
floor_top_mean_weighted = np.average(floor_top_l, weights = 1.0/floor_top_err**2)
error_floor_top_weighted = np.sqrt(1.0/np.sum(1.0/np.square(floor_top_err)))

print(floor_bottom_mean_weighted, error_floor_bottom_weighted)
print(floor_top_mean_weighted, error_floor_top_weighted)

#floor to center of mass
floor_center_of_mass = floor_top_mean_weighted - floor_bottom_mean_weighted
error_floor_center_of_mass = np.sqrt((error_floor_top_weighted)**2 + (-error_floor_bottom_weighted)**2)
print(floor_center_of_mass, error_floor_center_of_mass)
```

```python
# weighted pendulum measurements

floor_to_bottom = [
    (24.8, 0.15),
    (25.1, 0.2),
    (25.5, 0.1),
    (25.2, 0.12),
    (25.1, 0.18)
]

floor_to_top = [
    (49.8, 0.2),
    (50.1, 0.2),
    (50.5, 0.1),
    (49.99, 0.12),
    (49.9, 0.18)
]

pendulum_length_laser = [
    (22.257, 1*10e-3),
    (22.259, 1*10e-3),
    (22.257, 1*10e-3),
    (22.256, 1*10e-3),
    (22.257, 1*10e-3)
]

pendulum_length_tape = [
    (22.258, 0.8*10e-02),
    (22.268, 1.0*10e-02),
    (22.257, 0.7*10e-02),
    (22.249, 1.0*10e-02),
    (22.255, 1.0*10e-02)
]

# Subtract 0.1 from the first value in each tuple
adjusted_pendulum_lengths_laser = [
    (round(length - 0.1, 5), error)
    for length, error in pendulum_length_laser
]

adjusted_pendulum_lengths_tape = [
    (round(length - 0.1, 5), error)
    for length, error in pendulum_length_tape
]

lengths_tape = np.array([data[0] for data in adjusted_pendulum_lengths_tape])
errors_lengths_tape = np.array([data[1] for data in adjusted_pendulum_lengths_tape])

lengths_laser = np.array([data[0] for data in adjusted_pendulum_lengths_laser])
errors_lengths_laser = np.array([data[1] for data in adjusted_pendulum_lengths_laser])

length_tape_mean_weighted = np.average(
    lengths_tape, weights=1.0 / errors_lengths_tape**2
)
error_length_tape_weighted = np.sqrt(
    1.0 / np.sum(1.0 / np.square(errors_lengths_tape))
)

length_laser_mean_weighted = np.average(
    lengths_laser, weights=1.0 / errors_lengths_laser**2
)
error_length_laser_weighted = np.sqrt(
    1.0 / np.sum(1.0 / np.square(errors_lengths_laser))
)

print(length_tape_mean_weighted, error_length_tape_weighted)
print(length_laser_mean_weighted, error_length_laser_weighted)

# adjusted for floor to center of mass
adjusted_length_tape_mean_weighted = length_tape_mean_weighted - floor_center_of_mass
error_adjusted_length_tape_weighted = np.sqrt(
    error_length_tape_weighted**2 + error_floor_center_of_mass**2
)

adjusted_length_laser_mean_weighted = length_laser_mean_weighted - floor_center_of_mass
error_adjusted_length_laser_weighted = np.sqrt(
    error_length_laser_weighted**2 + error_floor_center_of_mass**2
)

print('tape: ', adjusted_length_tape_mean_weighted, error_adjusted_length_tape_weighted)
print('laser: ', adjusted_length_laser_mean_weighted, error_adjusted_length_laser_weighted)
```

```python
import numpy as np
from scipy.stats import chi2

def weighted_mean(x, sigma):
    w = 1.0 / sigma**2
    mean = np.sum(w * x) / np.sum(w)
    error = np.sqrt(1.0 / np.sum(w))
    return mean, error

def chi2_test(x, sigma, mean):
    chi2_val = np.sum((x - mean)**2 / sigma**2)
    dof = len(x) - 1
    p_val = 1.0 - chi2.cdf(chi2_val, dof)
    return chi2_val, dof, p_val

def rms_uncertainty(x):
    N = len(x)
    rms = np.sqrt(np.sum((x - np.mean(x))**2) / (N - 1))
    return rms

lengths_tape = np.array([l for l, e in adjusted_pendulum_lengths_tape])
errors_tape = np.array([e for l, e in adjusted_pendulum_lengths_tape])

Ltape, sigma_Ltape = weighted_mean(lengths_tape, errors_tape)
chi2_tape, dof_tape, p_tape = chi2_test(lengths_tape, errors_tape, Ltape)

print(f"L_tape = {Ltape:.5f} ± {sigma_Ltape:.5f} m")
print(f"χ² = {chi2_tape:.2f}, dof = {dof_tape}, p = {p_tape:.2f}")

lengths_laser = np.array([l for l, e in adjusted_pendulum_lengths_laser])
errors_laser = np.array([e for l, e in adjusted_pendulum_lengths_laser])

Llaser_w, sigma_Llaser_w = weighted_mean(lengths_laser, errors_laser)
chi2_laser, dof_laser, p_laser = chi2_test(lengths_laser, errors_laser, Llaser_w)

print(f"L_laser (weighted) = {Llaser_w:.5f} ± {sigma_Llaser_w:.5f} m")
print(f"χ² = {chi2_laser:.2f}, dof = {dof_laser}, p = {p_laser:.2e}")

Llaser_mean = np.mean(lengths_laser)
sigma_Llaser_rms = rms_uncertainty(lengths_laser)

print(f"L_laser (RMS) = {Llaser_mean:.5f} ± {sigma_Llaser_rms:.5f} m")

L_values = np.array([Ltape, Llaser_mean])
L_errors = np.array([sigma_Ltape, sigma_Llaser_rms])

L_final, sigma_L_final = weighted_mean(L_values, L_errors)
chi2_L, dof_L, p_L = chi2_test(L_values, L_errors, L_final)

print(f"L = {L_final:.5f} ± {sigma_L_final:.5f} m")
print(f"χ² = {chi2_L:.5f}, dof = {dof_L}, p = {p_L:.5f}")
```
