```python
#weighted pendulum measurements

lengths_tape = np.array([data[0] for data in adjusted_pendulum_lengths_tape])
errors_lengths_tape = np.array([data[1] for data in adjusted_pendulum_lengths_tape])
lengths_laser = np.array([data[0] for data in adjusted_pendulum_lengths_tape])
errors_lengths_laser = np.array([data[1] for data in adjusted_pendulum_lengths_tape])

length_tape_mean_weighted = np.average(lengths_tape, weights = 1.0/errors_lengths_tape**2)
error_length_tape_weighted = np.sqrt(1.0/np.sum(1.0/np.square(errors_lengths_tape)))
length_laser_mean_weighted = np.average(lengths_laser, weights = 1.0/errors_lengths_laser**2)
error_length_laser_weighted = np.sqrt(1.0/np.sum(1.0/np.square(errors_lengths_laser)))
length_tape_mean_weighted, error_length_tape_weighted, length_laser_mean_weighted, error_length_laser_weighted
```
