```python
#weighted measurements form floor

floor_bottom_l = np.array([data[0]*0.001 for data in floor_to_bottom])
floor_bottom_err = np.array([data[1]*0.001 for data in floor_to_bottom])
floor_top_l = np.array([data[0]*0.001 for data in floor_to_top])
floor_top_err = np.array([data[1]*0.001 for data in floor_to_top])

floor_bottom_mean_weighted = np.average(floor_bottom_l, weights = 1.0/floor_bottom_err**2)
error_floor_bottom_weighted = np.sqrt(1.0/np.sum(1.0/np.square(floor_bottom_err)))
floor_top_mean_weighted = np.average(floor_top_l, weights = 1.0/floor_top_err**2)
error_floor_top_weighted = np.sqrt(1.0/np.sum(1.0/np.square(floor_top_err)))

print(floor_bottom_mean_weighted, error_floor_bottom_weighted)
print(floor_top_mean_weighted, error_floor_top_weighted)

#floor to center of mass
floor_center_of_mass = floor_top_mean_weighted - floor_bottom_mean_weighted
error_floor_center_of_mass = np.sqrt((error_floor_top_weighted)**2 + (-error_floor_bottom_weighted)**2)
print(floor_center_of_mass, error_floor_center_of_mass)
```
